```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_sal = 0
highest_name = ""

for name,salary in employees.items():
    if salary>highest_sal:
        highest_sal = salary
        highest_name = name
print(highest_name ,"has a salary of:",highest_sal)
```

```python
numbers = [45, 22, 89, 10, 66]
minimum = numbers[0]
maximum = numbers[0]
for number in numbers:
    if number > maximum:
        maximum = number
    if number < minimum:
        minimum = number
print("List:",numbers)
print("max:",maximum)
print("min:",minimum)
```

```python
attendance = ["P", "P", "A", "P", "P"]
days_present = 0
for record in attendance:
    if record == "P":
        days_present += 1
total_days = len(attendance)
percentage = (days_present / total_days) * 100
print("Attendance Percentage:", percentage)
```
